```python

from IPython.display import display, Image
image_path = "wakeupcat.jpg"
display(Image(filename=image_path))
import io
import os

# Imports the Google Cloud client library
from google.cloud import vision

# Instantiates a client
client = vision.ImageAnnotatorClient()

# The name of the image file to annotate
file_name = os.path.abspath(image_path)

# Loads the image into memory
with io.open(file_name, 'rb') as image_file:
    content = image_file.read()

image = vision.Image(content=content)

# Performs label detection on the image file
response = client.label_detection(image=image)
labels = response.label_annotations

print('Labels:')
for label in labels:
    print(label.description)
```

```python
def detect_faces(path):
    """Detects faces in an image."""
    from google.cloud import vision
    import io
    client = vision.ImageAnnotatorClient()

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image    = vision.Image(content=content)
    response = client.face_detection(image=image)
    faces    = response.face_annotations

    # Names of likelihood from google.cloud.vision.enums
    likelihood_name = ('UNKNOWN', 'VERY_UNLIKELY', 'UNLIKELY', 'POSSIBLE',
                       'LIKELY', 'VERY_LIKELY')
    print('Faces:')
    result = []
    for face in faces:
        print('anger: {}'.format(likelihood_name[face.anger_likelihood]))
        print('joy: {}'.format(likelihood_name[face.joy_likelihood]))
        print('surprise: {}'.format(likelihood_name[face.surprise_likelihood]))

        vertices = ([(vertex.x, vertex.y)
                    for vertex in face.bounding_poly.vertices])

        print('face bounds: {}'.format(vertices))
        result.append(vertices)
    if response.error.message:
        raise Exception(
            '{}\nFor more info on error messages, check: '
            'https://cloud.google.com/apis/design/errors'.format(
                response.error.message))
    return result
```

```python
image_file = 'file7az0bh7emx2mflq99gg.jpg'
display(Image(filename=image_file))
vertices = detect_faces(image_file)
```

```python
import matplotlib
from matplotlib import pyplot
```

```python
img = matplotlib.image.imread(image_file)
figure, ax = pyplot.subplots(1)

ax.imshow(img) 

for vertice in vertices:
    [p1, p2, p3, p4 ] = vertice
    height            = p4[1]-p1[1]
    width             = p2[0]-p1[0]
    rect = patches.Rectangle(p1,height,width, edgecolor='r', facecolor="none")
    ax.add_patch(rect)
```

```python
# Imports the Google Cloud client library
from google.cloud import language_v1
```

```python
# Instantiates a client
client = language_v1.LanguageServiceClient()
```

## 4.1) sentiments

```python
def sentiment_analysis(text):
    document = language_v1.Document(
        content=text, 
        type_=language_v1.Document.Type.PLAIN_TEXT
    )
    json_request = {"document": document}
    # Detects the sentiment of the text

    analysis = client.analyze_sentiment(request=json_request)
    sentiments = analysis.document_sentiment
    return analysis
```

```python
# The text to analyze
text = "Thanks you !"
```

```python
sentiment_analysis(text)
```

```python
sentiment_analysis(text)
```

# 4.2) entities

```python
def entities_analysis(text):
    document = language_v1.Document(
        content=text, 
        type_=language_v1.Document.Type.PLAIN_TEXT
    )
    json_request = {"document": document}
    # Detects the sentiment of the text

    response = client.analyze_entities(
                 document=document,
                 encoding_type='UTF32',
             )
    
    return response
```

```python
text = "life is like google : you just have to know what you are searching for."
```

```python
sentiment_analysis(text)
```

```python
entities_analysis(text)
```

```python
text = 'Michelangelo Caravaggio, Italian painter, is known for "The Calling of Saint Matthew".'
response = entities_analysis(text)
for i, entity in enumerate(response.entities):
        print('=' * 20, i, '=' * 20)
        print('         name: {0}'.format(entity.name))
        print('         type: {0}'.format(entity.type))
        print('     metadata: {0}'.format(entity.metadata))
        print('     salience: {0}'.format(entity.salience))
        print(response.entities)
```

# 5) Google translate

```python
from google.cloud import translate
```

```python
def translate_text(text="Hello, world!", source_language_code="en-US",target_language_code="es" ):
    project_id  = "helical-sanctum-353821"
    client      = translate.TranslationServiceClient()
    location    = "global"
    parent      = f"projects/{project_id}/locations/{location}"
    response    = client.translate_text(
                    request={
                        "parent": parent,
                        "contents": [text],
                        "mime_type": "text/plain",
                        "source_language_code": source_language_code,
                        "target_language_code": target_language_code,
                    }
                )

    for translation in response.translations:
        print("Translated text: {}".format(translation.translated_text))
```

```python
original = "Your time is limited, so don’t waste it living someone else’s life."

translate_text(original)
```
